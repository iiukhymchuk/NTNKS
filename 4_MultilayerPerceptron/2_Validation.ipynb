```python
import joblib
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)

tr_features = pd.read_csv('../train_features.csv')
tr_labels = pd.read_csv('../train_labels.csv', header=None)
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    stds = results.cv_results_['std_test_score']
    sorted_results = sorted(zip(means, stds, results.cv_results_['params']), key=lambda x: x[0], reverse=True)
    for mean, std, params in sorted_results:
        print('{} (+/-{}) for {}'.format(round(mean, 3), round(std * 2, 3), params))
```

```python
mlp = MLPClassifier()
parameters = {
    'solver': ['lbfgs', 'adam'],
    'hidden_layer_sizes': [(10, 1), (20, 1), (50, 1), (100, 1)],
    'activation': ['relu', 'tanh', 'logistic'],
    'learning_rate': ['constant', 'invscaling', 'adaptive']
}

cv = GridSearchCV(mlp, parameters, cv=5)
cv.fit(tr_features, tr_labels.values.ravel())

print_results(cv)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

tr_features = pd.read_csv('../train_features.csv')
tr_labels = pd.read_csv('../train_labels.csv', header=None)

val_features = pd.read_csv('../val_features.csv')
val_labels = pd.read_csv('../val_labels.csv', header=None)

te_features = pd.read_csv('../test_features.csv')
te_labels = pd.read_csv('../test_labels.csv', header=None)
```

```python
model1 = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(20, 1), learning_rate='adaptive', activation='logistic', max_iter=10000)
model1.fit(tr_features, tr_labels.values.ravel())

model2 = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(10, 1), learning_rate='invscaling', activation='tanh', max_iter=10000)
model2.fit(tr_features, tr_labels.values.ravel())

model3 = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(100, 1), learning_rate='invscaling', activation='logistic', max_iter=10000)
model3.fit(tr_features, tr_labels.values.ravel())
```

```python
for mdl in [model1, model2, model3]:
    y_pred = mdl.predict(val_features)
    accuracy = round(accuracy_score(val_labels, y_pred), 3)
    precision = round(precision_score(val_labels, y_pred), 3)
    recall = round(recall_score(val_labels, y_pred), 3)
    print('Layers: {} / Learning rate: {} / Activation {} -- A: {} / P: {} / R: {}'.format(
        mdl.hidden_layer_sizes,
        mdl.learning_rate,
        mdl.activation,
        accuracy,
        precision,
        recall))
```

```python
for mdl in [model1, model2, model3]:
    y_pred = mdl.predict(te_features)
    accuracy = round(accuracy_score(te_labels, y_pred), 3)
    precision = round(precision_score(te_labels, y_pred), 3)
    recall = round(recall_score(te_labels, y_pred), 3)
    print('Layers: {} / Learning rate: {} / Activation {} -- A: {} / P: {} / R: {}'.format(
            mdl.hidden_layer_sizes,
            mdl.learning_rate,
            mdl.activation,
            accuracy,
            precision,
            recall))
```
